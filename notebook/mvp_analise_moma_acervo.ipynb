A variável `Nationality` contém múltiplos valores por registro, refletindo obras associadas a mais de um artista. Para permitir a análise, os dados foram transformados de modo que cada nacionalidade fosse considerada individualmente.

Nesta análise, a nacionalidade é contabilizada com base nas ocorrências nos registros das obras, e não em uma base deduplicada de artistas únicos. Assim, os resultados indicam a presença relativa das nacionalidades nas obras do acervo.

- remoção de registros inválidos  
- agregação das ocorrências  

Foram também criadas variáveis intermediárias para organizar o fluxo de transformação:
- `Nationality_list`: nacionalidades separadas individualmente

A padronização foi realizada a partir dos valores observados após a limpeza, convertendo gentilícios em países correspondentes. Casos ambíguos ou não diretamente associados a Estados nacionais foram simplificados quando possível ou mantidos como não mapeados.

```python
# preparando a variável Gender para análise

# criando dataset derivado
df_gender = df[['Gender']].copy()

# removendo valores nulos
df_gender = df_gender.dropna(subset=['Gender'])

# removendo parênteses
df_gender['Gender_clean'] = (
    df_gender['Gender']
    .str.replace('(', '', regex=False)
    .str.replace(')', '', regex=False)
    .str.strip()
)

# separando múltiplos valores
df_gender['Gender_list'] = df_gender['Gender_clean'].str.split()

# explode na lista para que cada gênero ocupe uma linha
df_gender = df_gender.explode('Gender_list')

# limpando espaços
df_gender['Gender_list'] = df_gender['Gender_list'].str.strip()

# removendo vazios
df_gender = df_gender[df_gender['Gender_list'] != '']

# visualizando o resultado
df_gender.head()
```

### 5.1.8 Distribuição temporal das obras (Decade)

Nesta análise é observada a distribuição temporal das obras a partir da variável derivada *Decade*. A partir da padronização da variável *Date*, é possível identificar os períodos históricos com maior presença no acervo.

### Preparação da variável temporal

A variável `Date` apresenta registros em formatos variados, o que dificulta a análise direta da distribuição temporal das obras. Para viabilizar a análise, foi criada uma base derivada (`df_date`) com a extração do ano de produção a partir dos registros disponíveis.

A partir do ano identificado, foi construída a variável `Decade`, permitindo uma representação temporal mais agregada e consistente. Registros sem informação válida de ano foram mantidos como ausentes.

Essa transformação facilita a leitura de padrões históricos no acervo, preservando a integridade dos dados originais.

```python
# criando a variável derivada Decade

# criando o dataset derivado com a coluna necessária
df_date = df[['Date']].copy()

# padronizando a coluna Date como texto
df_date['Date'] = df_date['Date'].astype(str).str.strip()

# extraindo o primeiro ano de quatro dígitos encontrado na string
df_date['Year'] = df_date['Date'].str.extract(r'(\d{4})')

# convertendo a coluna Year para numérica
df_date['Year'] = pd.to_numeric(df_date['Year'], errors='coerce')

# criando a coluna Decade a partir do ano extraído
df_date['Decade'] = (
    (df_date['Year'] // 10) * 10
).astype('Int64').astype(str) + 's'

# substituindo valores inválidos gerados a partir de ausentes
df_date.loc[df_date['Year'].isna(), 'Decade'] = pd.NA

# visualizando amostra do resultado
df_date[['Date', 'Year', 'Decade']].head(10)
```

```python
# gráfico de distribuição temporal das obras (Decade)

# contagem de valores
df_decade_count = (
    df_date['Decade']
    .dropna()
    .value_counts()
    .reset_index(name='Count')
    .rename(columns={'index': 'Decade'})
)

# criando coluna auxiliar numérica para ordenar cronologicamente
df_decade_count['Decade_order'] = (
    df_decade_count['Decade']
    .str.replace('s', '', regex=False)
    .astype(int)
)

# ordenando pelas décadas
df_decade_count = df_decade_count.sort_values('Decade_order')

# criando o gráfico
fig = px.bar(
    df_decade_count,
    x='Decade',
    y='Count',
    title='Distribuição temporal das obras por década de criação (MoMA)',
    color='Count',
    color_continuous_scale='Viridis_r',
    labels={
        'Decade': 'Década',
        'Count': 'Número de obras'
    }
)

# ajustes visuais
fig.update_layout(
    margin=dict(t=60, l=20, r=20, b=20),
    coloraxis_showscale=False
)

fig.show()
```

### Análise da variável Decade (Distribuição temporal das obras)

A distribuição temporal das obras revela forte concentração ao longo do século XX, com destaque para as décadas de 1960, 1970 e 1940, que apresentam os maiores volumes de registros. A década de 1960 se sobressai como o período mais representado, com mais de 23 mil ocorrências.

Observa-se crescimento consistente ao longo do século XX, especialmente entre as décadas de 1920 e 1970, seguido por uma leve redução nas décadas posteriores. Ainda assim, períodos mais recentes, como os anos 1990, 2000 e 2010, mantêm presença relevante no acervo.

As décadas anteriores ao século XX apresentam volumes reduzidos, indicando menor representação de obras mais antigas. Registros anteriores ao século XIX são residuais, sugerindo concentração predominante na produção moderna e contemporânea.

De modo geral, a distribuição evidencia:
- concentração no século XX, especialmente entre 1920 e 1970  
- presença contínua de obras contemporâneas nas décadas recentes  
- baixa representatividade de períodos anteriores à modernidade  

Esses padrões são coerentes com o perfil institucional do MoMA, voltado à arte moderna e contemporânea, refletindo o recorte curatorial e a formação do acervo ao longo do tempo.

### 5.2.3 Distribuição das décadas de criação por tipo de obra (Decade × Classification)

Esta análise investiga como os períodos históricos estão representados na coleção do MoMA, considerando a distribuição dos tipos de obra de cada época.

A variável `Decade` representa o período de criação das obras e constitui o eixo principal da análise, enquanto `Classification` permite observar como diferentes tipos de produção artística contribuem para a composição de cada década.

A combinação dessas dimensões possibilita compreender como os períodos históricos se distribuem no acervo a partir da quantidade e diversidade de tipos de obra, evidenciando quais décadas concentram maior volume e quais meios artísticos estão mais presentes em cada período.

```python
# gráfico de evolução dos tipos de obra ao longo das décadas

# preparando os dados
df_dc = df[['Date', 'Classification']].copy()

# extraindo  década
df_dc['Date'] = df_dc['Date'].astype(str).str.strip()
df_dc['Year'] = df_dc['Date'].str.extract(r'(\d{4})')
df_dc['Year'] = pd.to_numeric(df_dc['Year'], errors='coerce')

df_dc['Decade'] = ((df_dc['Year'] // 10) * 10).astype('Int64').astype(str) + 's'
df_dc.loc[df_dc['Year'].isna(), 'Decade'] = pd.NA

# removendo nulos
df_dc = df_dc.dropna(subset=['Decade', 'Classification'])

# agregando valores
df_grouped = (
    df_dc
    .groupby(['Decade', 'Classification'])
    .size()
    .reset_index(name='Count')
)

# ordenando décadas
df_grouped['Decade_order'] = df_grouped['Decade'].str.replace('s', '', regex=False).astype(int)
df_grouped = df_grouped.sort_values('Decade_order')

decade_order = (
    df_grouped[['Decade', 'Decade_order']]
    .drop_duplicates()
    .sort_values('Decade_order')['Decade']
    .tolist()
)

# criando o gráfico
fig = px.bar(
    df_grouped,
    x='Decade',
    y='Count',
    color='Classification',
    title='Tipos de obra no acervo do MoMA por década de criação',
    labels={
        'Decade': 'Década de criação da obra',
        'Count': 'Número de obras no acervo',
        'Classification': 'Tipo de obra'
    },
    category_orders={'Decade': decade_order}
)

# ajustes visuais
fig.update_layout(
    barmode='stack',
    margin=dict(t=60, l=20, r=20, b=20),
    xaxis_title='Década de criação da obra',
    yaxis_title='Número de obras no acervo'
)

fig.show()
```

A análise evidencia como diferentes períodos históricos estão representados no acervo do MoMA a partir da distribuição dos tipos de obra em cada década de criação.

Observa-se forte concentração de obras produzidas ao longo do século XX, com destaque para as décadas de 1930, 1940, 1950 e, principalmente, 1960, que apresenta o maior volume do acervo. Esse padrão reforça o recorte do museu voltado à arte moderna, com maior densidade de obras produzidas nesse período.

Períodos anteriores ao século XX apresentam baixa representatividade, com ocorrências pontuais, reforçando a concentração do acervo em momentos associados à arte moderna e contemporânea.

### 5.2.4 Distribuição das obras por departamento segundo a década de criação (Decade × Department)

Esta análise investiga como as obras de diferentes décadas de criação se distribuem entre os departamentos do MoMA, a partir do cruzamento entre as variáveis `Decade` e `Department`.

A combinação dessas variáveis permite identificar padrões de concentração por período histórico, evidenciando em quais departamentos se concentram as obras de determinadas décadas.

Essa abordagem contribui para compreender a composição do acervo em termos institucionais, destacando como produções de diferentes épocas estão organizadas entre os departamentos e possíveis ênfases curatoriais na formação da coleção.

```python
# gráfico de evolução dos departamentos ao longo das décadas (heatmap)

# preparando os dados
df_dd = df[['Date', 'Department']].copy()

# padronizando a coluna Date como texto
df_dd['Date'] = df_dd['Date'].astype(str).str.strip()

# extraindo o primeiro ano válido
df_dd['Year'] = df_dd['Date'].str.extract(r'(\d{4})')
df_dd['Year'] = pd.to_numeric(df_dd['Year'], errors='coerce')

# criando a coluna de década
df_dd['Decade'] = ((df_dd['Year'] // 10) * 10).astype('Int64').astype(str) + 's'
df_dd.loc[df_dd['Year'].isna(), 'Decade'] = pd.NA

# removendo registros sem década ou departamento
df_dd = df_dd.dropna(subset=['Decade', 'Department'])

# agregando dados
df_grouped = (
    df_dd
    .groupby(['Department', 'Decade'])
    .size()
    .reset_index(name='Count')
)

# ordenando as décadas
df_grouped['Decade_order'] = df_grouped['Decade'].str.replace('s', '', regex=False).astype(int)
df_grouped = df_grouped.sort_values('Decade_order')

decade_order = (
    df_grouped[['Decade', 'Decade_order']]
    .drop_duplicates()
    .sort_values('Decade_order')['Decade']
    .tolist()
)

# criando a matriz para o heatmap (pivot)
df_pivot = df_grouped.pivot(
    index='Department',
    columns='Decade',
    values='Count'
).fillna(0)

# reordenando colunas (décadas)
df_pivot = df_pivot[decade_order]

# criando o heatmap
fig = px.imshow(
    df_pivot,
    aspect='auto',
    title='Distribuição das obras por departamento segundo a década de criação (MoMA)',
    color_continuous_scale='Blues',
    labels=dict(
        x='Década de criação da obra',
        y='Departamento',
        color='Número de obras'
    )
)

# ajustes visuais
fig.update_layout(
    margin=dict(t=60, l=20, r=20, b=20),
    coloraxis_colorbar=dict(
        title='Nº de obras'
    )
)

fig.show()
```

### Análise da distribuição das obras por departamento segundo a década de criação (Decade × Department)

A distribuição das obras por departamento segundo a década de criação evidencia concentração do acervo em determinados períodos artísticos, com destaque consistente para *Drawings & Prints* ao longo do século XX.

Observa-se que *Drawings & Prints* reúne elevado número de obras de diversas décadas, com pico na década de 1960 (mais de 15 mil registros), mantendo volumes expressivos também de obras das décadas de 1920, 1940, 1950, 1970, 1980, 1990 e 2000. Esse padrão reforça o papel central das práticas gráficas e editoriais na composição do acervo.

As décadas de 1930 a 1950 também se destacam em *Architecture & Design*, com volumes elevados de obras desses períodos, sugerindo a presença de conjuntos documentais ligados à arquitetura e ao design.

Já *Photography* concentra maior volume de obras produzidas a partir da década de 1970, indicando maior representatividade de produções desse período no acervo.

De modo geral, a análise evidencia:
- predominância de obras de múltiplas décadas em *Drawings & Prints*  
- forte presença de períodos específicos, especialmente entre 1930 e 1950, em *Architecture & Design*  
- maior concentração de obras da segunda metade do século XX em *Photography*  

Esses padrões indicam que a distribuição das obras entre os departamentos varia conforme o período de criação, refletindo tanto características da produção artística quanto critérios institucionais de organização e incorporação ao acervo.

## 6.2 Limpeza dos dados

A etapa de limpeza dos dados tem como objetivo garantir a consistência e a qualidade do dataset selecionado para modelagem, removendo possíveis inconsistências que possam comprometer análises futuras.

Nesta fase, são avaliados aspectos como duplicidade de registros, presença de valores inválidos e integridade das variáveis selecionadas. A limpeza é realizada de forma criteriosa, buscando preservar o máximo de informação relevante, ao mesmo tempo em que elimina elementos que possam introduzir distorções nos resultados.

```python
# limpeza dos dados

# criando cópia de trabalho para limpeza
df_ml_clean = df_ml.copy()

# removendo espaços extras em variáveis categóricas
categorical_cols = [
    'Classification',
    'Department',
    'Country',
    'Gender_clean',
    'Decade'
]

for col in categorical_cols:
    df_ml_clean[col] = df_ml_clean[col].astype('string').str.strip()

# verificando quantidade de registros duplicados
duplicated_count = df_ml_clean.duplicated().sum()
print(f'Registros duplicados encontrados: {duplicated_count}')

# removendo duplicatas, se existirem
df_ml_clean = df_ml_clean.drop_duplicates()

# exibindo dimensões do dataset após limpeza
print(f'Dimensões após limpeza: {df_ml_clean.shape}')

# visualizando amostra do dataset limpo
df_ml_clean.head()
```

A remoção dessas duplicidades foi adotada como estratégia de redução de redundância, permitindo trabalhar com um conjunto mais enxuto de perfis distintos de obras. Essa abordagem é particularmente adequada para técnicas como clusterização, nas quais o foco está na identificação de padrões, e não na frequência absoluta dos registros.

### 6.3.3 Gender

Na variável *Gender*, valores ausentes foram padronizados como "unknown". Adicionalmente, categorias minoritárias foram agrupadas, visando reduzir a fragmentação e tornar a variável mais adequada para análise e modelagem.

```python
# padronizando e tratando valores ausentes em Gender_clean

def clean_gender(g):
    # Tratar ausentes
    if pd.isna(g):
        return 'unknown'

    # Padronizar texto
    g = str(g).lower().strip()

    # Regras de agrupamento
    if g == 'male':
        return 'male'
    elif g == 'female':
        return 'female'
    elif g in ['non-binary', 'gender non-conforming', 'transgender woman',
               'female (transwoman', 'male (trans? ftm?']:
        return 'non-binary/trans'
    else:
        return 'unknown'

df_ml_missing['Gender_clean'] = df_ml_missing['Gender_clean'].apply(clean_gender)

# Inspeção rápida
df_ml_missing['Gender_clean'].value_counts(dropna=False)
```

## 6.4 Transformação de variáveis categóricas (encoding)

Como os algoritmos de machine learning operam sobre dados numéricos, foi necessário transformar as variáveis categóricas selecionadas em um formato compatível com modelagem.

Para isso, foi aplicado o método de *one-hot encoding*, que converte cada categoria em uma variável binária, preservando a informação original sem impor relações de ordem artificial entre os valores.

Essa estratégia é especialmente adequada para variáveis nominais, como *Classification*, *Department*, *Country*, *Gender_clean* e *Decade*. A variável *YearAcquired*, por já se encontrar em formato numérico, foi mantida sem transformação.

O resultado dessa etapa é um dataset integralmente numérico, pronto para ser utilizado em algoritmos de clusterização e outras técnicas de aprendizado de máquina.


```python
# transformação de variáveis categóricas (encoding)

# definindo variáveis categóricas
categorical_cols = [
    'Classification',
    'Department',
    'Country',
    'Gender_clean',
    'Decade'
]

# aplicando one-hot encoding
# drop_first=False mantém todas as categorias, o que é útil
# para análise exploratória e clusterização
df_ml_encoded = pd.get_dummies(
    df_ml,
    columns=categorical_cols,
    drop_first=False
)

# visualizar estrutura resultante
print("Dimensões do dataset após encoding:", df_ml_encoded.shape)

df_ml_encoded.head()
```

### Resultado da transformação de variáveis categóricas (encoding)

Após a aplicação do one-hot encoding, o dataset passou a apresentar 18.323 registros e 197 variáveis, refletindo a expansão das variáveis categóricas em múltiplas colunas binárias.

Cada categoria das variáveis originais (*Classification*, *Department*, *Country*, *Gender_clean* e *Decade*) foi transformada em uma coluna indicadora, assumindo valores booleanos (*True* ou *False*) conforme a presença daquela característica em cada registro.

A variável *YearAcquired* foi mantida em formato numérico, sendo a única variável contínua do conjunto.

O aumento no número de colunas é esperado nesse tipo de transformação e representa a decomposição das categorias em atributos independentes, permitindo que algoritmos de machine learning interpretem corretamente as informações sem assumir relações de ordem entre os valores.

De forma geral, o dataset resultante encontra-se integralmente numérico (considerando a representação booleana como binária), estruturado e adequado para aplicação de técnicas de clusterização e outras abordagens de aprendizado de máquina.

## 6.5 Padronização e normalização

Após a etapa de encoding, todas as variáveis do dataset encontram-se em formato numérico. No entanto, essas variáveis apresentam escalas distintas, o que pode impactar negativamente algoritmos baseados em distância, como os de clusterização.

Em particular, a variável *YearAcquired* possui valores significativamente maiores do que as variáveis binárias geradas pelo one-hot encoding, podendo exercer influência desproporcional na formação dos agrupamentos.

Para mitigar esse efeito, foi aplicada a padronização dos dados por meio do método *StandardScaler*, que transforma as variáveis para uma distribuição com média zero e desvio padrão unitário.

Essa transformação garante que todas as variáveis contribuam de forma equilibrada para o modelo, permitindo a identificação de padrões mais consistentes e representativos no conjunto de dados.

```python
# padronização e normalização

# inicializar o scaler
scaler = StandardScaler()

# aplicando padronização
df_ml_scaled = scaler.fit_transform(df_ml_encoded)

# convertendo de volta para DataFrame
# (mantendo nomes das colunas)
df_ml_scaled = pd.DataFrame(
    df_ml_scaled,
    columns=df_ml_encoded.columns
)

# visualizando resultado
print("Dimensões após padronização:", df_ml_scaled.shape)

df_ml_scaled.head()
```

### Resultado da padronização e normalização

Após a aplicação da padronização por meio do *StandardScaler*, o dataset manteve suas dimensões originais (18.323 registros e 197 variáveis), indicando que não houve alteração estrutural, apenas transformação na escala dos dados.

As variáveis passaram a apresentar valores centrados em torno de zero, com distribuição ajustada em termos de variância, conforme esperado nesse tipo de transformação. Isso pode ser observado na substituição dos valores originais por números positivos e negativos de menor magnitude, refletindo sua posição relativa em relação à média.

Essa etapa é fundamental para garantir que variáveis originalmente em escalas distintas — como *YearAcquired* e as variáveis binárias geradas pelo encoding — contribuam de forma equilibrada para os cálculos de distância utilizados em algoritmos de clusterização.

## 6.6 Exploração inicial com K-Means

Como etapa exploratória complementar, foi aplicada uma técnica de aprendizado não supervisionado com o algoritmo K-Means (clusterização), com o objetivo de identificar possíveis padrões e agrupamentos entre as obras do acervo.

A análise foi realizada a partir do dataset previamente tratado e padronizado, garantindo que todas as variáveis estivessem em formato numérico e comparável. O uso do K-Means permite agrupar registros com características semelhantes, sem a necessidade de uma variável alvo pré-definida.

Essa abordagem busca investigar se existem perfis distintos de obras dentro da coleção, considerando simultaneamente variáveis como classificação, departamento, nacionalidade, gênero e período histórico.

Embora não tenha caráter conclusivo, a análise demonstra o potencial da aplicação de técnicas de aprendizado não supervisionado em acervos culturais, podendo apoiar diferentes frentes de atuação, como:
- novas formas de organização e segmentação do acervo
- desenvolvimento de exposições temáticas
- identificação de lacunas de representatividade
- apoio à leitura estratégica da coleção

Os resultados devem ser interpretados como uma exploração inicial, que pode ser aprofundada em aplicações futuras mais robustas.

### 6.6.1 Determinação do número de clusters (método do cotovelo)

Antes de aplicar o algoritmo K-Means, é necessário definir o número adequado de clusters (K). Para isso, utilizamos o método do cotovelo, que consiste em testar diferentes valores de K e observar a variação da inércia do modelo.

A inércia representa a soma das distâncias entre os pontos e o centro de seus respectivos clusters. À medida que aumentamos o número de clusters, a inércia tende a diminuir. No entanto, a partir de certo ponto, essa redução passa a ocorrer de forma menos significativa.

O ponto de inflexão da curva (conhecido como "cotovelo") indica um equilíbrio entre simplicidade e capacidade de representação dos dados, sendo utilizado como referência para a escolha do número de clusters mais adequado.

```python
# teste do cotovelo (Elbow Method)

inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_ml_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(6,4))
plt.plot(K_range, inertia, marker='o')
plt.title('Método do Cotovelo')
plt.xlabel('Número de clusters (K)')
plt.ylabel('Inércia')
plt.show()
```

### 6.6.2 Escolha do número de clusters

A partir da aplicação do método do cotovelo, observa-se que a redução da inércia ocorre de forma mais acentuada até aproximadamente K = 5. A partir desse ponto, a curva apresenta uma diminuição mais gradual, indicando ganhos marginais na compactação dos clusters.

Esse comportamento sugere que valores acima de 5 não proporcionam melhorias significativas na qualidade do agrupamento, ao mesmo tempo em que aumentam a complexidade da interpretação.

Dessa forma, optou-se por utilizar K = 5 como número de clusters para a etapa de modelagem, buscando um equilíbrio entre capacidade de segmentação e interpretabilidade dos resultados.

```python
# escolhendo do número de clusters

# número de clusters definido para o modelo final
k_optimal = 5

print(f'Número de clusters selecionado: {k_optimal}')
```

### 6.6.3 Aplicação do modelo K-Means

Com o número de clusters definido, o algoritmo K-Means é aplicado ao dataset previamente padronizado.

Nesta etapa, o modelo atribui cada registro a um dos cinco clusters, com base na similaridade entre suas características. O resultado é incorporado ao dataset de modelagem, permitindo examinar posteriormente a distribuição e o perfil de cada agrupamento.

Essa etapa marca a primeira aplicação efetiva de uma técnica de aprendizado não supervisionado no projeto, com foco exploratório e interpretativo.


```python
# clusterização aplicando o modelo K-Means

# inicializando o modelo
kmeans = KMeans(
    n_clusters=k_optimal,
    random_state=42,
    n_init=10
)

# treinando o modelo no dataset padronizado
kmeans.fit(df_ml_scaled)

# adicionando os clusters ao dataset de modelagem
df_ml['Cluster'] = kmeans.labels_

# inspeção inicial do resultado
df_ml.head()
```

### 6.6.4 Distribuição dos clusters

Após a aplicação do modelo K-Means, é importante analisar a distribuição dos registros entre os clusters formados.

Essa verificação permite identificar se os agrupamentos estão equilibrados ou se há concentração excessiva em determinados clusters, o que pode indicar padrões dominantes ou possíveis limitações na segmentação.

A análise do tamanho dos clusters é um passo inicial para compreender a estrutura dos agrupamentos antes de avançar para a interpretação de seus perfis.

```python
# distribuição dos clusters

# contagem de registros por cluster
cluster_counts = df_ml['Cluster'].value_counts().sort_index()

print('Distribuição dos clusters:\n')
print(cluster_counts)

# visualização simples

plt.figure(figsize=(6,4))
cluster_counts.plot(kind='bar')

plt.title('Distribuição dos clusters')
plt.xlabel('Cluster')
plt.ylabel('Quantidade de registros')
plt.xticks(rotation=0, ha='center')

plt.show()
```

### 6.6.5 Perfil dos clusters

Após verificar o tamanho dos clusters, a próxima etapa consiste em examinar o perfil de cada agrupamento.

Para isso, são observadas as distribuições das principais variáveis categóricas dentro de cada cluster, com foco em *Classification*, *Department* e *Country*. Essa leitura permite identificar quais características predominam em cada grupo e quais perfis de obra tendem a se formar.

```python
# perfil dos clusters - Country

top_country = (
    df_ml.groupby('Cluster')['Country']
    .value_counts()
    .groupby(level=0)
    .head(5)
)

print("Top 5 Country por cluster:\n")
print(top_country)
```
